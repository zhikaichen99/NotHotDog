```python
directories = ['train', 'test']
img_classes = ['hot_dog', 'not_hot_dog']
hot_dog_counter = 0
not_counter = 0

for directory in directories:
    for img_class in img_classes:
        folder = './data/{}/{}/*.jpg'.format(directory, img_class)
        img_paths = glob.glob(folder)
        
        for path in img_paths:
            full_size_image = Image.open(path)
            im = full_size_image.resize((224,224))
            if img_class == 'hot_dog':
                im.save('./data/{}_{}.jpg'.format(img_class, hot_dog_counter))
                hot_dog_counter = hot_dog_counter + 1
            else:
                im.save('./data/{}_{}.jpg'.format(img_class, not_counter))
                not_counter = not_counter + 1
```
